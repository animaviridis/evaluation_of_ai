```python
# how many browsers has a participant used
browsers_known = main_data[keys['search_engines'][1]]
browsers_known.head()
```

```python
c = [len(s.split(';')) for s in browsers_known]
l = len(c)
l2 = l//2
print(c)
print(sorted(c))
print(f"min: {min(c)}, max: {max(c)}, mean: {sum(c)/len(c)}, median: {sorted(c)[l2] if l%2 else sum(sorted(c)[l2:l2+2])/2}")
```
