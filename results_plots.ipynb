```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import aux_functions as aux
import plotting_functions as plots

import matplotlib
%matplotlib inline
```

```python
main_data = pd.read_csv('data/main_survey.csv')
```

```python
main_data.head()
```

```python
# divide questions
k = list(main_data.keys())
keys = dict()
keys['general'] = k[2:6]
keys['search_engines'] = k[6:9]
keys['phrase'] = k[9:10]
keys['tasks'] = k[10:18]
keys['clustering'] = k[18:20]
keys['yippy'] = k[20:23] + k[25:26]
keys['remarks_yippy'] = k[23:24]
keys['remarks_eval'] = [k[26], k[28]]
keys['eval'] = [k[27]]
```

```python
fig = plots.make_plot(main_data, keys['general'], type='pie', fig_kwargs=dict(figsize=(9,6)), show=False)
fig.subplots_adjust(left=0.0, hspace=0.25)
```

```python
fig = plots.make_plot(main_data, keys['yippy'], type='bar', fig_kwargs=dict(figsize=(9,6)), break_labels=False, show=False)
fig.subplots_adjust(left=0.0, hspace=0.5, wspace=0.2)
```

```python
fig = plots.make_plot(main_data, keys['tasks'], type='bar', fig_kwargs=dict(figsize=(12,18)), show=False,
                      nrows=len(keys['tasks'])//2)
fig.subplots_adjust(left=0.0, hspace=0.5, wspace=0.2)
```
